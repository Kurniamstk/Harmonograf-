# **Harmonograph** 
**Harmonograf** adalah alat mekanis yang menggunakan pendulum untuk membuat gambar geometris. 

Harmonograf membuat figurnya menggunakan gerakan pendulum yang teredam. Pergerakan pendulum teredam dijelaskan oleh persamaan :
image.png

f = Frekuensi 

p = fase

A = Amplitudo

d = redaman 

t = waktu 

Jika bandul itu dapat bergerak pada sumbu (dalam bentuk lingkaran atau elips) karena prinsip superposisi, gerak batang yang terhubung ke dasar bandul sepanjang satu sumbu akan dijelaskan oleh persamaan :

image.png

Harmonografi tipikal memiliki dua pendulum yang bergerak sedemikian rupa, dan pena yang digerakkan oleh dua batang tegak lurus yang terhubung ke pendulum ini. Oleh karena itu, jalur gambar harmoniografi dijelaskan oleh persamaan parametrik :

image.png

**Ilustrasi Membuat Harmonograph Dengan Python**

Python dapat menerjemahkan persamaan ini menjadi grafik yang meniru homonigraf. 

```python
#library aljabar dan grafis
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameter yang digunakan
n = 1000000
t = np.logspace(np.log10(10),np.log10(500),n)

# Silakan merubah parameter ini
A = [  1, 6, 1.5, 1.5 ]
d = [ .004, .001, .002, .0015 ]
f = [   3, 1, 4, 2.5 ]

# Membuat pasangan x dan y
x = A[0]*np.sin(t*f[0])*np.exp(-d[0]*t) + A[1]*np.sin(t*f[1])*np.exp(-d[1]*t)
y = A[2]*np.sin(t*f[2])*np.exp(-d[2]*t) + A[3]*np.sin(t*f[3])*np.exp(-d[3]*t)

# Menampilkan plot nya
plt.plot(x,y,'r',linewidth=.1)
plt.axis('off')
plt.show()
```
